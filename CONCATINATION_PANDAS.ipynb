```python
#---------concatination---------


import pandas as pd
d1=[[34,56],[89,64]]
d2=[[76,70],[26,61]]
c=['A','B']
df=pd.DataFrame(d1,columns=c)
df1=pd.DataFrame(d2,columns=c)
result=[df,df1]
d3=pd.concat(result)
print(d3)
```

```python
#------concationation by using names--------


import pandas as pd
d1=[['kohli','dhoni'],['dravid','sachin']]
d2=[['rohit','jadeja'],['shami','umesh']]
c=['A','B']
df=pd.DataFrame(d1,columns=c)
df1=pd.DataFrame(d2,columns=c)
result=[df,df1]
d3=pd.concat(result,ignore_index=True)
print(d3)
```

```python
#---------------------------------------------#


import pandas as pd
d1=[['kohli','dhoni'],['dravid','sachin']]
d2=[['rohit','jadeja'],['shami','umesh']]
c=['A','B']
z=['x','y']
df=pd.DataFrame(d1,columns=c)          #----by default it taking the axis=1----
df1=pd.DataFrame(d2,columns=z)
result=[df,df1]
d3=pd.concat(result,ignore_index=True)
print(d3)
```
