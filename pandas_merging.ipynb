```python
##--------------------creating dataframes-------------------------------------------

import pandas as pd
d1={'id':[1,2,3,4,5,6],
'name':['pradhan','venu','madhurima','nireekasha','shafi','veeru'],
'subject':['english','java','html','python','c','dot net']}
d2={'id':[11,12,13,14,15,16],
    'name':['srinu','sumanth','neelima','daniel','arjun','veeru'],
    'subject':['java','html','cpp','python','c','dotnet']}
df=pd.DataFrame(d1)
df1=pd.DataFrame(d2)
print(df)
print()
print(df1)
```

```python
#---------------innermerging---------------------------


df3=pd.merge(df,df1,on='subject',how='inner')
print(df3)
```

```python
#--------------rightmerging-----------------


import pandas as pd
d1={'id':[1,2,3,4,5,6],
'name':['pradhan','venu','madhurima','nireekasha','shafi','veeru'],
'subject':['english','java','html','python','c','dot net']}
d2={'id':[11,12,13,14,15,16],
    'name':['srinu','sumanth','neelima','daniel','arjun','veeru'],
    'subject':['java','html','cpp','python','c','dotnet']}
df=pd.DataFrame(d1)
df1=pd.DataFrame(d2)
rightjoin=pd.merge(df,df1,on='subject',how='right')
print(rightjoin)
```

```python
#--------------leftmerging---------------



import pandas as pd
d1={'id':[1,2,3,4,5,6],
'name':['pradhan','venu','madhurima','nireekasha','shafi','veeru'],
'subject':['english','java','html','python','c','dot net']}
d2={'id':[11,12,13,14,15,16],
    'name':['srinu','sumanth','neelima','daniel','arjun','veeru'],
    'subject':['java','html','cpp','python','c','dotnet']}
df=pd.DataFrame(d1)
df1=pd.DataFrame(d2)
leftjoin=pd.merge(df,df1,on='subject',how='left')
print(leftjoin)
```

```python
#------------outermerging----------------


import pandas as pd
d1={'id':[1,2,3,4,5,6],
'name':['pradhan','venu','madhurima','nireekasha','shafi','veeru'],
'subject':['english','java','html','python','c','dot net']}
d2={'id':[11,12,13,14,15,16],
    'name':['srinu','sumanth','neelima','daniel','arjun','veeru'],
    'subject':['java','html','cpp','python','c','dotnet']}
df=pd.DataFrame(d1)
df1=pd.DataFrame(d2)
fulljoin=pd.merge(df,df1,on='subject',how='outer')
print(fulljoin)
```
